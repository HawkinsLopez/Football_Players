### Displaying 10 the Tallest Players

```python
#using searborn plot Height in cm First name Last name Country of birth Position City of Birth
sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(x="height_in_cm",y="first_name",data=height)
sns.barplot(x="height_in_cm",y="last_name",data=height)
plt.show()
```

### Player With High Market Value in Europe

```python
df[['highest_market_value_in_eur','first_name','last_name','current_club_name']].sort_values(ascending=False,by='highest_market_value_in_eur',ignore_index=True)[:5]
```

### vizualizing the top 5 players with highest market value in euro

```python
#using seaborn displaying names and the higst market value in eur
sns.barplot(x=df['highest_market_value_in_eur'].value_counts().head(10).index, y=df['highest_market_value_in_eur'].value_counts().head(10).values,palette='RdGy_r')
plt.title('Top 10 higest market vale Players')
plt.xlabel('highest_market_value_in_eur')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.figure(figsize=(10,5))
plt.show()
```
